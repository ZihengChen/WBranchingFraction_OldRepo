# 1. Load Data

```python
MCttsig = pd.read_csv("MCttsig.csv").drop('Unnamed: 0', axis=1)
MCttbkg = pd.read_csv("MCttbkg.csv").drop('Unnamed: 0', axis=1)
MCbkg = pd.read_csv("MCbkg.csv").drop('Unnamed: 0', axis=1)
Data  = pd.read_csv("dat.csv").drop('Unnamed: 0', axis=1)

MCbkg.drop('label', axis=1, inplace=True)
MCttbkg.drop('label', axis=1, inplace=True)
MCttsig.drop('label', axis=1, inplace=True)
Data.drop('label', axis=1, inplace=True)
```

# 2. Prediction

```python
net = torch.load("nuDNN.pt")
dflist = [MCbkg,MCttbkg,MCttsig,Data]
print("load net")
net = net.cpu()
NFeature = 69+8

for df in dflist:#range(len(mydflist)):
    temp = df.reset_index(drop=True)
    temp['nntag'] = np.ones(len(temp.index)).astype(int)
    tempset    = MyDataset(temp.as_matrix(),NFeature)
    temploader = DataLoader(tempset, batch_size=tempset.__len__(),shuffle=False, num_workers=1)
    tempiter = iter(temploader).next()
    tempinputs   = tempiter["feature"]
    templabels   = tempiter["label"]
    tempoutputs  = net(Variable(tempinputs))
    temppredicts = torch.max(tempoutputs.data, 1)[1] # return likelihood, predict
    temppredicts = temppredicts.numpy()
    #df = df.reindex()
    df["nntag"] = pd.Series(temppredicts,index=df.index)
```

# 3. Making Plots

```python
cata=2
MCbkg_ = MCbkg[MCbkg.nntag==cata]
MCttbkg_ = MCttbkg[MCttbkg.nntag==cata]
MCttsig_ = MCttsig[MCttsig.nntag==cata]
Data_ = Data[Data.nntag==cata]
SF = np.array([1,0.008,18.935,4.227,1.297,0.0387])
```

```python
    v,a,b,step,xl =  'lepton2_pt',0,150,5,'lepton2_pt [GeV]'
    mybin = np.arange(a,b,step)
    
    hbkg,_ = np.histogram(MCbkg_[v],bins=mybin,  weights=0.0387*2*np.ones(len(MCbkg_)))
    hbttsig,_ = np.histogram(MCttsig_[v],bins=mybin,  weights=0.0387*2*np.ones(len(MCttsig_)))
    httbkg,_ = np.histogram(MCttbkg_[v],bins=mybin,  weights=0.0387*2/0.4*np.ones(len(MCttbkg_)))
    hdata,_ = np.histogram(Data_[v],bins=mybin)
    
    center = mybin[:-1]
    plt.bar(center, hbttsig, width=step, bottom=None,lw=0,color="#54c3d7",alpha=0.6)
    plt.bar(center, httbkg , width=step, bottom=hbttsig,lw=0,color="#49feec",alpha=0.6)
    hmctotal = hbttsig+httbkg
    
    
    center = mybin[:-1]+step/2
    plt.errorbar(center,hmctotal,yerr=hmctotal**0.5,color="k",fmt='none',lw=200/mybin.size, mew=0, alpha=0.2)
    # 2. show data
    plt.errorbar(center,hdata-hbkg,yerr=hdata**0.5,color="k",fmt='.', label="data")
    
    # 4. plot settings
    plt.legend(fontsize=10,loc="upper right")
    plt.text(0.04*b+0.96*a, 3600, r'CMS $preliminary$',style="italic",fontsize="15",fontweight='bold')
        
    plt.grid()
    plt.ylim(1,4000)
    plt.xlim(a,b)
    #plt.ylim(1,10*h.max())
    #plt.yscale('log')
    plt.xlabel(xl)
    plt.title("L=35.7/fb (13TeV)",loc="right")
    plt.savefig("../plotnudnn/subtract/_{}.png".format(v))
   
```
