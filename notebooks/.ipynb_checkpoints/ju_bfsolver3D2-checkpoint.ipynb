```python
tb     = BFCalc3D_Toolbox()
# defining bfCalc
df_acc = tb.IO_LoadAccTableIntoDf()
xs     = 832 + 35.6
usetag = "1b"

a_mm   = df_acc.loc[(df_acc.sel=="mumu")  & (df_acc.tag==usetag),'acc'].values[0]
a_mt   = df_acc.loc[(df_acc.sel=="mutau") & (df_acc.tag==usetag),'acc'].values[0]
a_em   = df_acc.loc[(df_acc.sel=="emu")   & (df_acc.tag==usetag),'acc'].values[0]
a_mh   = df_acc.loc[(df_acc.sel=="mu4j")  & (df_acc.tag==usetag),'acc'].values[0]


bf = BFCalc3D_ThreeSelectorRatios(a_mm, a_mt, a_em, a_mh, xs=xs, lumin=35700)
```

```python
r_list, Be_list, Bm_list = [],[],[]
for i in range(1000):
    xmm,xmt,xem = bf.PredictXs_SmearOfSubN()
    r,Be,Bm  = bf.SovleBF(xmm,xmt,xem)
    r_list.append(r)
    Be_list.append(Be)
    Bm_list.append(Bm)
    
r_list  = np.array(r_list)
Be_list = np.array(Be_list)
Bm_list = np.array(Bm_list)
```
