```python
def getmets(i):
    metx = np.cos(df.metphi[i])*df.met[i]
    mety = np.sin(df.metphi[i])*df.met[i]

    mu    = r.TLorentzVector(df.lepton1_px[i],df.lepton1_py[i],df.lepton1_pz[i],
                             df.lepton1_energy[i])
    mu_   = r.TLorentzVector(df.lepton2_px[i],df.lepton2_py[i],df.lepton2_pz[i],
                             df.lepton2_energy[i])
    b     = r.TLorentzVector(df.jet1_px[i],df.jet1_py[i],df.jet1_pz[i],df.jet1_energy[i])
    b_    = r.TLorentzVector(df.jet2_px[i],df.jet2_py[i],df.jet2_pz[i],df.jet2_energy[i])

    twobjets = (b,b_)
    twomuons = (mu,mu_)
    totalmet = (metx,mety)
    NutrinoSolver = doubleNeutrinoSolutions(twobjets,twomuons,totalmet)
    
    s = NutrinoSolver.nunu_s

    minsumdeltar = 10000
    for slt in range(2):
        
        temp = s[slt][0]
        nu   = r.TLorentzVector(temp[0],temp[1],temp[2],np.linalg.norm(temp) )
        temp = s[slt][1]
        nu_  = r.TLorentzVector(temp[0],temp[1],temp[2],np.linalg.norm(temp) )
        sumdeltar = nu.DeltaR(mu)+nu_.DeltaR(mu_)
        if sumdeltar<minsumdeltar:
            minsumdeltar = sumdeltar
            result = s[slt]
    return result[0],result[1]

def getmt(ptmu,ptnu):
    temp = np.linalg.norm(ptmu)*np.linalg.norm(ptnu) - np.dot(ptmu,ptnu)
    temp = 2*temp
    return temp**0.5
```

```python
met1,met2   = np.ones([0,4]),np.ones([0,4])

for i in range(len(df)):
    if (i%10000==0):
        print(i)
    nu1,nu2 = getmets(i)
    ptmu1 = np.array(df.loc[i,["lepton1_px","lepton1_py"]])
    ptmu2 = np.array(df.loc[i,["lepton2_px","lepton2_py"]])
    
    aa = np.r_[nu1,getmt(ptmu1,nu1[0:2])]
    aa = aa[None,:]
    bb = np.r_[nu2,getmt(ptmu2,nu2[0:2])]
    bb = bb[None,:]
    met1 = np.r_[met1,aa]
    met2 = np.r_[met2,bb]
    
np.save("../data/{}_met1".format(filename),met1)
np.save("../data/{}_met2".format(filename),met2)
```
