# 4. Template Fitting

```python
from utility_templatefit import *
PlotHistogramWithFit("lepton2_pt",0,150,5,"lepton2_pt [GeV]", 
                     df_list,label_list,color_list,
                     "../plot/{}/".format(selection),
                     lock=np.array([1,1,0,0,0,0])
                    )
```

```python
mchist, mcerr2, datahist = GetMCDataHistAndErr("lepton2_pt",0,150,5,df_list)
alpha_init = np.ones(mchist.shape[0])

def chisqured(adjust,mchist, mcerr2, datahist):
    sigma2  = (datahist+np.dot(adjust**2,mcerr2))
    mask    = sigma2>0
    cost    = (datahist-np.dot(adjust,mchist))**2
    cost    = np.sum(cost[mask]/sigma2[mask])
    
    penalty = (adjust-np.ones(adjust.size))**2
    penalty = np.dot(1e6*penalty,np.array([1,1,0,0,0,0]))
    return cost + np.sum(penalty)
result = minimize(chisqured, x0=alpha_init,
                  args=(mchist, mcerr2, datahist),
                  bounds=[(0,10)]*6,
                  method='SLSQP')
```

```python
_,__ =PlotHistogram("lepton2_pt",0,150,5,"lepton2_pt [GeV]", 
                    df_list,label_list,color_list,
                    adjust = result.x
                   )
```
