```python
from pylab import *
plt.rc("figure",facecolor="w",figsize=(6,4))
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('../data/BWl_pdg2017.csv')
```

```python
df.label[1]
```

```python
plt.rc("figure",facecolor="w",figsize=(6,6))

plt.axvline(10.86,color='k', linestyle="--",lw=0.5)
low = 10.86-0.09
up =  10.86+0.09
plt.fill([low,low,up,up], [-20,20,20,-20],'k',alpha=0.4) 

# tau
plt.text(11.7,4,r'$Br(W\to\tau\nu)$')
plt.errorbar(df.BWt[0:4], np.arange( 0, 4,1), xerr=df.u_BWt[0:4], fmt='.',color='C1')
plt.errorbar(df.BWt[4], 4, xerr=df.u_BWt[4], fmt='o',color='C1',markersize=5)
for i in np.arange(0,5,1):
    plt.text(9,i,'{:5}'.format(df.label[i]))
    plt.text(13 ,i,r'${:5.2f}\pm{:4.2f}$'.format(df.BWt[i],df.u_BWt[i]))
plt.axhline(5, linestyle="-",color='k',lw=0.5)


# mu
plt.text(11.7,10,r'$Br(W\to\mu\nu)$')
plt.errorbar(df.BWm[0:4], np.arange( 6, 10,1), xerr=df.u_BWm[0:4], fmt='.',color='C2')
plt.errorbar(df.BWm[4], 10, xerr=df.u_BWm[4], fmt='o',color='C2',markersize=5)
for i in np.arange(0,5,1):
    plt.text(9,i+6,'{:5}'.format(df.label[i]))
    plt.text(13 ,i+6,r'${:5.2f}\pm{:4.2f}$'.format(df.BWm[i],df.u_BWm[i]))
plt.axhline(11, linestyle="-",color='k',lw=0.5)


# e
plt.text(11.7,16,r'$Br(W\to e\nu)$')
plt.errorbar(df.BWe[0:4], np.arange(12,16,1), xerr=df.u_BWe[0:4], fmt='.',color='C3')
plt.errorbar(df.BWe[4], 16, xerr=df.u_BWe[4], fmt='o',color='C3',markersize=5)
for i in np.arange(0,5,1):
    plt.text(9,i+12,'{:5}'.format(df.label[i]))
    plt.text(13 ,i+12,r'${:5.2f}\pm{:4.2f}$'.format(df.BWe[i],df.u_BWe[i]))
    
plt.xlim(8.5,14.5)
plt.ylim(-0.5,17)
plt.xlabel(r'$Br(W\to l\nu)$'+'[%]')
plt.yticks([])
plt.title('PDG 2017',loc='right')
plt.savefig("pdg2017.png",dpi=300)
```
